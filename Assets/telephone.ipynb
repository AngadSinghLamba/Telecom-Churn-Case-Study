## Problem Statement

The telecom industry faces a high churn rate (15-25%) which is costly due to the higher expense of acquiring new customers compared to retaining existing ones. The project aims to address this by building a predictive model to identify customers at high risk of churn, specifically focusing on high-value customers in the Indian and Southeast Asian prepaid markets. The model will utilize customer-level data over four months, defining churn based on usage patterns. The ultimate goal is to not only predict churn but also understand its key indicators to inform customer retention strategies.

```python
import pandas as pd
```

### Step 1: Inspecting the Dataframe

```python
df_tel=pd.read_csv("telecom_churn_data.csv")
pd.set_option('display.max_columns', None)
```

```python
df_tel.head(2)
```

#### Filter high-value customers

```python
# Calculate the average recharge amount for the first two months
df_tel['avg_rech_amt_6_7'] = (df_tel['total_rech_amt_6'] + df_tel['total_rech_amt_7']) / 2

# Calculate the 70th percentile of the average recharge amount
recharge_threshold = df_tel['avg_rech_amt_6_7'].quantile(0.7)

# Filter high-value customers
df_tel = df_tel[df_tel['avg_rech_amt_6_7'] > recharge_threshold].copy()

# Display the number of high-value customers
print(f"Number of high-value customers: {df_tel.shape[0]}")
```

```python
df_tel.reset_index(inplace=True)
```

```python
df_tel.drop("index",axis=1,inplace=True)
```

```python
df_tel.shape
```

#### Tagging Churned Customers


###### We define churn as a customer having no incoming/outgoing calls 


```python
# Create the 'churn' column based on the specified conditions
df_tel['churn'] = ((df_tel['total_ic_mou_9'] + df_tel['total_og_mou_9'] + 
                    df_tel['vol_2g_mb_9'] + df_tel['vol_3g_mb_9']) == 0).astype(int)

# Drop columns related to the churn phase
cols_to_drop = [col for col in df_tel.columns if col.endswith('_9')]
df_tel.drop(cols_to_drop, axis=1, inplace=True)

# Print the count of churned and non-churned customers
print("Churned customers count:", df_tel['churn'].sum())
print("Non-churned customers count:", (df_tel['churn'] == 0).sum())
```
